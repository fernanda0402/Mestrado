```
import sys
import warnings
sys.path.append("analytic-sl-solutions-main/src/circular_source")

import Arcs
warnings.filterwarnings("ignore")
```

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import MaxNLocator
import mpmath
from matplotlib import cm

mpmath.mp.dps = 10
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 25
```

Let's define a huge number of points for plotting in the interval $[0,2\pi]$

```
angles = np.linspace(0, 2*np.pi, 50000)
```

To obtain the shape of the source in the source plane (circular image), we can use either $\beta_+$ or $\beta_-$ functions:

```
#par√¢metros do sistema
phi_0 = 0
r0 = 0.25
s = 0.5
theta_e = 1
```

```
source_positions_betaplus = Arcs.betaplus(phi=angles,\
                phi_0=phi_0, r0=r0, s=s);
source_positions_betaminus = Arcs.betaminus(phi=angles,\
                phi_0=phi_0, r0=r0, s=s);
```

Since this is the radial coordinate, we can convert into cartesian coordinates (we might use either source_positions_betaplus or source_positions_betaminus because we are evaluating over all possible angles):

```
x_source = source_positions_betaplus * np.cos(angles)
y_source = source_positions_betaplus * np.sin(angles)
```

```
plt.figure(figsize=(5,5))
plt.plot(x_source, y_source)
plt.grid(True);
```

To obtain the arcs on the image plane, we might use Arcs.thetaplus_plus and Arcs.thetaplus_minus or Arcs.thetaminus_plus and Arcs.thetaminus_minus. You can check and see that they are equal because of range of angles.

```
image_positions_external_arcs_point_lens = Arcs.thetaplus_plus(phi=angles,\
                phi_0=phi_0, r0=r0, s=s, theta_e=theta_e)
image_positions_internal_arcs_point_lens = Arcs.thetaplus_minus(phi=angles,\
                phi_0=phi_0, r0=r0, s=s, theta_e=theta_e)
image_positions_external_arcs_sis_lens = Arcs.xmais(phi=angles, r0=r0, s=s)
image_positions_internal_arcs_sis_lens = Arcs.xmenos(phi=angles, r0=r0, s=s)
```

```
x_image_point_lens_external_arcs = image_positions_external_arcs_point_lens * np.cos(angles)
y_image_point_lens_external_arcs = image_positions_external_arcs_point_lens * np.sin(angles)
x_image_point_lens_internal_arcs = image_positions_internal_arcs_point_lens * np.cos(angles)
y_image_point_lens_internal_arcs = image_positions_internal_arcs_point_lens * np.sin(angles)
x_image_sis_lens_external_arcs = image_positions_external_arcs_sis_lens * np.cos(angles)
y_image_sis_lens_external_arcs = image_positions_external_arcs_sis_lens * np.sin(angles)
x_image_sis_lens_internal_arcs = image_positions_internal_arcs_sis_lens * np.cos(angles)
y_image_sis_lens_internal_arcs = image_positions_internal_arcs_sis_lens * np.sin(angles)
```

```
fig = plt.figure(figsize=(15,20))
ax = fig.add_subplot(111)
ax.plot(x_image_point_lens_external_arcs, y_image_point_lens_external_arcs, label='Outer Arc PL')
ax.plot(x_image_point_lens_internal_arcs, y_image_point_lens_internal_arcs, label='Inner Arc PL')
ax.plot(x_image_sis_lens_external_arcs, y_image_sis_lens_external_arcs, label='Outer Arc SIS')
ax.plot(x_image_sis_lens_internal_arcs, y_image_sis_lens_internal_arcs, label='Inner Arc SIS')
ax.plot(x_source, y_source, label='Source')
ax.set_yticks([])
ax.legend(fontsize=15)
ax.set_aspect('equal')
# ax.set_aspect('auto')
plt.show()
```
